```python
import math
import random
import datetime

question = [
    ("what is the square root of 81?",math.sqrt(81)),
    ("what is 2 raised to power 6?",math.pow(2,6)),
    ("what is factorial of 5?",math.factorial(5)),
    ("what is log base 10 of 1000?",math.log10(1000))

]
random.shuffle(question)

print("welcome to the math quiz game!")
print("you have 5 question.let begin...\n")

score =0
start_time = datetime.datetime.now()

for i,(q,ans) in enumerate(question, 1):
    print(f"question {i}:")
    print(q)
    user_answer = input("your answer: ") # Added input to get user answer
    try:
        if abs(float(user_answer) - float(ans)) <0.01: # Added missing parenthesis and colon
               print("correct!\n")
               score+= 1
        else:
               print(f"wrong! correct answer = {ans}\n")
    except ValueError: # Changed to ValueError for invalid float conversion
           print(f" Invalid input! Correct answer = {ans}\n")

end_time = datetime.datetime.now()
time_taken = end_time - start_time

print("quiz over!")
print(f"your score: {score}/{len(question)}")
print(f"time taken: {time_taken.second} seconds")
```
