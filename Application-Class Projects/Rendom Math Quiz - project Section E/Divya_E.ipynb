```python
import math
import random
import datetime

problems = [ ("What is the square root of 16 ?", math.sqrt(16)),
            ("What is 3 raised to the power 7?", math.pow(3, 7)),
             ("What is the factorial of 6?", math.factorial(6)),
             ("What is tan(45)?", round(math.tan(math.radians(45)), 2)),
             ("What is log base 10 of 200?", math.log10(200))]

random.shuffle(problems)
print("Welcome to the Maths Quiz Game!")
print("You have 5 questions. Let's begin...\n")
score=0
start_time = datetime.datetime.now()

for i, (q, ans) in enumerate(problems, 1):
  print(f"P {i}: {q}")
  user_answer = input("Your answer: ")

  try:
    if abs(float(user_answer) - float(ans)) < 0.01:
      print("Correct!\n")
      score +=1
    else:
      print(f"Wrong! Correct Answer= {ans}\n")

  except:
      print(f"Invaild input! Correct Answer = {ans}\n")

end_time = datetime.datetime.now()
time_taken = end_time - start_time
print("Quiz Over!")
print(f"Your Score: {score} / {len(problems)}")
print(f"Time Taken: {time_taken.seconds} seconds")
```
